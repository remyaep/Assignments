```python
def find_the_difference(s, t):
    """This sfunction finds teh difference in characters between 2 strings"""
    for i in s:
        if i in t:
            t= t.replace(i,'',1)
    return t

find_the_difference("abcd", "abcde")
```

```python
def count_datatypes(*args):
    """This function accepts unlimited arguments, check and count how
    many data types are in those arguments. Finally return the total in a list"""
    order=[int, str, bool, list, tuple, dict]
    type_args=[]
    for i in range(len(args)):
        type_args.append(type(args[i]))

    count=[type_args.count(order[i]) for i in range(len(order))]
    return count

count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6])
```

```python
def fib_str(length,fib_list):
    """This function creates afibonacci series of the 2 characters given"""
    a=fib_list[0]
    b=fib_list[1]
    o_string = [a,b]
    for i in range(length-2):
        a,b=b,b+a
        o_string.append(b)
    return o_string
    
fib_str(6, ["n", "k"])
```

```python
def ones_threes_nines(n):
    """This function returns the the number of nines, threes, and ones required to form the number"""
    nines=n//9
    out = "nines:"+str(nines)+", "
    b=n%9
    threes=b//3
    out = out+"threes:"+str(threes)+", "
    ones=b%3
    out = out+"ones:"+str(ones)
    return out
    
ones_threes_nines(22)
```

```python
def fib(n):
    """This function returns the fibonacci number at the nth position"""
    a=1
    b=1
    fib_l=[a,b]
    for i in range(n-2):
        a,b=b,b+a
        fib_l.append(b)
        
    return fib_l[n-1]

fib(6)
```
